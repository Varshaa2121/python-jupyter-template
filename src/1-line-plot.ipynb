### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import matplotlib.pyplot as plt

def plot_line():
    try:
        # Generate x values from 0 to 10
        x = list(range(11))
        # Compute y values as the square of x
        y = [i ** 2 for i in x]
        
        # Ensure x and y are of the same length and contain only numbers
        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise ValueError("x and y must contain only numerical values.")
        
        # Create the plot
        plt.plot(x, y, label='y = x^2')
        
        # Customize the plot
        plt.title("Line Plot: y = x^2")  # Title of the plot
        plt.xlabel("x values")  # Label for x-axis
        plt.ylabel("y values")  # Label for y-axis
        plt.grid(True)  # Add grid
        plt.legend()  # Display legend
        
        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Call the function to generate the plot
plot_line()
```
